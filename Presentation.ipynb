<font size='5' face="Times New Roman"> **The whole process to filter the most common and important words** <font>

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from collections import Counter
import matplotlib.pyplot as plt
from wordcloud import WordCloud

#Preprocessing the text
def preprocess_text(text):
    #Tokenization
    tokens = word_tokenize(text)   
    #Lowercasing
    tokens = [token.lower() for token in tokens]
    #Removing stopwords -> especially important due to the large number of "irrelevant" words.
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [token for token in tokens if token not in stop_words]
    #Lemmatization
    lemmatizer = WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in filtered_tokens]
    return lemmatized_tokens

#Calling the user to input the text
user_text = input("Please enter a string of text: ")

#Preprocess user text -> using the Tokenization, Lowercasing, Stop Word Removal, and Lemmatization
#function.
preprocessed_user_text = preprocess_text(user_text)

#Join preprocessed text -> Remaking the sentence.
preprocessed_text_joined = ' '.join(preprocessed_user_text)

#Generate the word cloud with top 20 words
wordcloud = WordCloud(width=800, height=400, background_color='white', max_words=20).generate(preprocessed_text_joined)

#Display word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()

```

<font size='5' face="Times New Roman"> **Term Frequency, Inverse Document Frequency and scoring both** <font>

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS

def create_tfidf_table(text):
    #Converting the English stop words to a list to make it fit
    stop_words_list = list(ENGLISH_STOP_WORDS)

    #Create a vector with all the stop words removed.
    vectorizer = TfidfVectorizer(stop_words=stop_words_list)

    #Fitting this vector to the text that has been put in
    vectorizer.fit([text])

    #Creating a TF-IDF vector
    tfidf_vector = vectorizer.transform([text])

    #Getting the feature names
    feature_names = vectorizer.get_feature_names_out()

    #Creating the scores
    tfidf_scores = tfidf_vector.toarray().flatten()

    #IDF Scores
    idf_scores = vectorizer.idf_

    #TF Scores
    tf_scores = tfidf_vector.toarray().flatten() / idf_scores

    #Creating a panda dataframe to store the scores
    df = pd.DataFrame({
        'Word': feature_names,
        'TF-score': tf_scores,
        'IDF-score': idf_scores,
        'TF-IDF-score': tfidf_scores
    })

    #Sorting the scoring in a descending order
    df = df.sort_values(by='TF-IDF-score', ascending=False)

    #Displaying the top 20 words
    df_top_20 = df.head(20)

    return df_top_20

#The input
text = input("Please enter the text: ")

#Creating a table to make it easier to see the top scores
tfidf_table = create_tfidf_table(text)
print(tfidf_table)
```

```python
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.naive_bayes import MultinomialNB  # Or your preferred classifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
import joblib

#Function to read a txt file with a lot of emails for training and testing
def read_emails_from_files(ham_path, spam_path):
    try:
        with open(ham_path, "r") as ham_file:
            ham_content = ham_file.read()
            ham_emails = ham_content.split("From: ")[1:]
        with open(spam_path, "r") as spam_file:
            spam_content = spam_file.read()
            spam_emails = spam_content.split("From: ")[1:]
        emails = ham_emails + spam_emails
        labels = ["ham"] * len(ham_emails) + ["spam"] * len(spam_emails)
        return emails, labels
    except FileNotFoundError:
        print("Error, path not found")
        return None, None

#File paths for both sort of mails
ham_path = "/Users/sebasleen/Desktop/Spam detection /Train/HAM.txt"
spam_path = "/Users/sebasleen/Desktop/Spam detection /Train/SPAM.txt"

#Reading the emails
emails, labels = read_emails_from_files(ham_path, spam_path)

if emails is None: 
    exit()

#Preprocessing with tokenization, stop word removal, and part-of-speech tagging
nltk.download('punkt', 'averaged_perceptron_tagger', quiet=True) 

#Say you want to switch the language to Dutch, French or anything you can do so here
stop_words = set(stopwords.words("dutch")) 

emails_preprocessed = []
for email in emails:
    #Tokenization
    words = nltk.word_tokenize(email.lower())  # Lowercase and tokenize

    #Stop word removal (adjust stop words to the required language in the above part)
    words = [word for word in words if word not in stop_words]

    #Part of Speech tagging which is optional because it does not improve the model that much in this example
    tags = nltk.pos_tag(words)  # Enable if desired

    #Join preprocessed words back into a string to be able to read it as a whole
    preprocessed_email = " ".join(words)
    emails_preprocessed.append(preprocessed_email)

#Train and testing split
X_train, X_test, y_train, y_test = train_test_split(emails_preprocessed, labels, test_size=0.2, random_state=42)

#Feature extraction witht the TF-IDF classifiers
count_vect = CountVectorizer()  # Consider ngrams if relevant
X_train_counts = count_vect.fit_transform(X_train)
tfidf_transformer = TfidfTransformer()
X_train_tfidf = tfidf_transformer.fit_transform(X_train_counts)

#Train classifier using Naive Bayes
clf = MultinomialNB().fit(X_train_tfidf, y_train)

#Make predictions on the test set and see the results of the model (accuracy of the model)
y_pred = clf.predict(tfidf_transformer.transform(count_vect.transform(X_test)))

#Evaluate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

#Save the trained classifier and vectorizer to be able to run it in the script in the next line of code
joblib.dump(clf, "email_classifier_model.pkl")
joblib.dump(count_vect, "email_vectorizer.pkl")

print("Model trained and saved for later use!")
```

<font size='6' face='Times new Roman'>
The script to run on a txt file<font>

```python
#Load the trained classifier and vectorizer
clf = joblib.load("email_classifier_model.pkl")
vectorizer = joblib.load("email_vectorizer.pkl")

#The function used to predict whether it is a spam or a ham mail
def predict_with_confidence(email_text):
    email_counts = vectorizer.transform([email_text])
    prediction = clf.predict(email_counts)[0]

    #Calculate confidence (accuracy levels) based on predicted class (whether it is a spam or ham)
    if prediction == "ham":
        confidence = clf.predict_proba(email_counts)[0][0] * 100  
    else:
        confidence = clf.predict_proba(email_counts)[0][1] * 100  
    return prediction.upper(), confidence

#Function to read emails from a file
def read_emails_from_file(file_path):
    with open(file_path, "r") as file:
        emails = file.read().split("From: ")[1:]  # Skip the first empty entry
    return emails

#Main function
def main():
    file_path = input("Enter the path to the text file containing the emails as a txt file and I will check if it's spam or ham!: ")

    emails = read_emails_from_file(file_path)

    for idx, email in enumerate(emails, start=1):
        prediction, confidence = predict_with_confidence(email)
        print(f"Email {idx}: {prediction} - Confidence: {confidence:.2f}%")

#Run the main function
if __name__ == "__main__":
    main()
```
